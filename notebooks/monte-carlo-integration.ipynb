# Monte Carlo integration

```python
import numpy as np
import matplotlib.pylab as plt
```

## Problem and exact solution

```python
def f1(t):
    return np.sin(t)

N = (10, 100, 1000)
tmin = 0
tmax = np.pi

t = np.arange(tmin, tmax, 0.01)
x = f1(t)

plt.plot(x)
plt.show()
```

$\int _{0}^{\pi}{\sin {x}}\,dx=2$

## Simple numerical integration

```python
plt.figure(figsize=(12, 4))

for idx, n in enumerate(N):
    dt = (tmax - tmin) / n
    t = np.arange(tmin, tmax, dt)
    x = f1(t)
    result = (x * dt).sum()
    plt.subplot(1, 3, idx + 1)
    plt.title("S = {}, N = {}".format(round(result, 3), n))
    plt.bar(t, x)
plt.show()
```

## Monte Carlo Integration

```python
plt.figure(figsize=(12, 4))

for idx, n in enumerate(N):
    dt = (tmax - tmin) / n
    t = np.random.uniform(tmin, tmax, n)
    np.sort(t)
    x = f1(t)
    result = (x * dt).sum()
    plt.subplot(1, 3, idx + 1)
    plt.title("S = {}, N = {}".format(round(result, 3), n))
    plt.bar(t, x)
plt.show()
```
