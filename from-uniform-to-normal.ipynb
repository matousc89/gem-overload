# From uniform to normal (distribution)

```python
import numpy as np
import matplotlib.pylab as plt
```

## Uniform vs normal

```python
N = 100000
BINS = 100

x_u = np.random.uniform(low=0, high=1, size=N)
x_n = np.random.normal(0, 1, size=N)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(x_u, bins=BINS)
plt.title("Uniform")
plt.xlabel("Value (in bin range)")
plt.ylabel("Count (occurence)")

plt.subplot(1, 2, 2)
plt.hist(x_n, bins=BINS)
plt.title("Normal")
plt.xlabel("Value (in bin range)")
plt.ylabel("Count (occurence)")

plt.tight_layout()
plt.show()
```

## Sum of multiple uniformly random variables

```python
N = 100000
K, J = 2, 3
BINS = 100

data = np.random.uniform(low=0, high=1, size=(K * J,N))


plt.figure(figsize=(12, 8))

for j in range(J):
    for k in range(K):
        num = j * K + k + 1
        x = data[:num,:].sum(axis=0)
        plt.subplot(K, J, num)
        plt.hist(x, bins=BINS)
        plt.title("Sum of {} variables".format(num))
        plt.xlabel("Value (in bin range)")
        plt.ylabel("Count (occurence)")
        

plt.tight_layout()
plt.show()
```
